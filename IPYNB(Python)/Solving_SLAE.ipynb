```python
import numpy as np
from numpy import linalg

a=np.zeros([10,10])
f=np.zeros([10,1])

# заполнение матрицы и вектора по условию задачи
for i in range(10):
    for j in range(10):
        if i==j:
            a[i,j]=1
        else:
            a[i,j]=1/(i+j+2)
a1=a   

for i in range(10):
    f[i]=1/(i+1)
f1=f 

#****************************************** Метод Гаусса *****************************************************#    
    
# Прямой ход - получение верхеней треуголной матрицы
for i in range(9):
    for j in range(i+1,10):
        a[j]=a[i]*a[j,i]/a[i,i]-a[j]
        f[j]=f[i]*a[j,i]/a[i,i]-f[j]
        
# Получаем значения корня с диаганальных елементов начиная с x[9,9] - последовательно подставляем в верхние уравнения
x=np.zeros([10,1])
for i in range(9, -1, -1):
    x[i]=f[i]
    for j in range(i+1, 10):
        x[i]-=a[i,j]*x[j]
    x[i]/=a[i,i]  
print('\n*************** 1 ******************')
print('\nКорни СЛАО Методом Гаусса\n')
for i in range(10):
    print('X',i+1,'=',x[i])
    
v=a1.dot(x)-f1
print('\nВектор невязки:\n')
for i in range(10):
    print(' V',i+1,'=',v[i])

#****************************************** Метод ПВР *****************************************************#

x=np.zeros([10,1])    
for i in range(10):
    x[i]=i
e=0.0001
# 1<=w<=2 верхняя релаксация
w=1.5
x_new=x
a=a1
f=f1
#np.linalg.norm() -  ф-ия Евклидовой нормы 
while np.linalg.norm(a1.dot(x)-f1, ord=2) > e:
    for i in range(10):
        x_new[i]=f[i]
        for j in range(10):
            if i != j:
                x_new[i]-=a[i,j]*x_new[j]
        x_new[i]/=a[i,i]
    for i in range(10):
        x[i]=x[i]*(1-w)+w*x_new[i]
print('\n*************** 2 ******************')        
print('\nКорни СЛАО Методом ПВР-верхней релаксацией:\n')
for i in range(10):
    print('X',i+1,'=',x[i]) 
v=a1.dot(x)-f1
print('\nВектор невязки:\n')
for i in range(10):
    print(' V',i+1,'=',v[i])


#*************************************** Характерные значения входной матрицы **********************************************#

# нахождение массива собственных чисел и массива собственных векторов
c,v=linalg.eig(a1)
# нахождение максимального значения из  массива собственных чисел
L_max=max(c)
L_min=min(c)
print('\n*************** 3 ******************') 
print('\n Наибольшее собственное число L_max=',L_max,'\n','Наименьшее собственное число L_min=',L_min,'\n')    

print('\n*************** 4 ******************') 
# нахождение числа обусловленности матрицы
t=np.linalg.norm(a1)*np.linalg.norm(np.linalg.inv(a1))
print('Число обусловленности матрицы η=',t)

```
