```python
import numpy as np
import matplotlib.pyplot as plt
import math

#     2-  arctgx M2  
list_arctg=[]
n=-5
while n<=5:
    n+=0.1
    list_arctg.append(1/(1+n**(2)))
M2=max(list_arctg)
print('M2='+ str(M2))
#   arctgx
M0=np.pi/2
print('M0='+str(M0))
#    
h1=0.01
Eokr=(2*M0*10**(-16))/h1
Emash=M2*h1/2

#  h optimal   
h=2*np.sqrt((M0*Emash)/M2)
print('h optimal='+str(h))
#   
x=np.linspace(-5,5,100)
f1=1/(1+x**(2))
f2=(np.arctan(x)-np.arctan(x-h))/h-Eokr-Emash
#   
f3=f1-f2

#   
plt.grid()
plt.title("   ")
plt.plot(x,f1)
plt.show()

plt.grid()
plt.title("  ")
plt.plot(x,f2)
plt.show()

plt.grid()
plt.title("   ")
plt.plot(x, f1, color='b')
plt.plot(x, f2, color='g')
plt.show()

plt.grid()
plt.title(" ")
plt.plot(x,f3)
plt.show()
```
