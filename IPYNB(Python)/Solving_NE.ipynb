```python
# 1)
# Выполните отделение корней аналитически и найдите один из корней методами дихотомии и хорд
# с абсолютной погрешностью 0,0001.
# x**4 - 4x**3 - 8x**2 + 1=0
# Построим график функции и графически определим отрезок [a;b]
import numpy as np
import matplotlib.pyplot as plt

def border(n):
    print('\n')
    for i in range(108):
        if i==52:
            print('ЗАДАНИЕ №',n,end='')
        print('*',end='')
    print('\n')
border(1)
def func(x):
    return x**4 - 4*x**3 - 8*x**2 + 1
e=0.0001
x=np.linspace(3,8,500)
plt.grid()
plt.plot(x,func(x))
#plt.vlines(0, y.min(), y.max(),'b')
plt.hlines(0, 3, 8,'black')
plt.show()

# По графику определяем отрезок [5;6], на котором ф-ия обращается в нуль

print('\t','МЕТОД ДИХОТОМИИ \n','\n','[a;b] - отрезок, на котором ф-ия обращается в нуль \n')
#a=float(input('Введите значение начала отрезка a='))
#b=float(input('Введите значение конца отрезка b='))
#a1=a
#b1=b
a=5
b=6
print('\n')
# Найдем значение точки середины отрезка с и соотв-ие значение ф-ии
c=(a+b)/2
# Если f(a)*f(c)>0 , то корень лежит в области отрезка, тогда b=c ,иначе корень не лежит в области отрезка, тогда a=c и с - корень уравнения
iterations=0
while abs(func(c))>e:
    iterations+=1
    c=(a+b)/2
    print(iterations,'-ая итерация. Значение х=',c,'   f(x)=',func(c))
    if func(a)*func(c)<0:
        b=c
    else:
        a=c
res1=c
print('\n','Корень данного уравнения X=',c,'\n')
print(' Колличество итераций методом Дихотомии : ',iterations,'\n')
print('\t','МЕТОД ХОРД \n','\n')
# Построим график прямой соединяющий концы отрезка [a;b]
#a=a1
#b=b1
a=5
b=6
# Уравнение прямой соедин-ей две точки (x-x1)/(x2-x1)=(y-y1)/(y2-y1)
# y=((x-a)*(func(b)-func(a)))/(b-a) +func(a)
def straight(x):
    return ((x-a)*(func(b)-func(a)))/(b-a) +func(a)

x=np.linspace(3,8,500)
t=np.linspace(a,b,100)
plt.grid()
plt.title("График ф-ии y(x)= x**4 - 4*x**3 - 8*x**2 + 1 и прямой в первой итерации")
plt.plot(x,func(x))
plt.plot(t,straight(t),'r')
#plt.vlines(0, y.min(), y.max(),'b')
plt.hlines(0, 3, 8,'black')
plt.show()
# при с - x=0 , y=0
c= a - (func(a)*(b-a))/(func(b)-func(a))
iterations=0
while abs(func(c))>e:
    iterations+=1
    c= a - (func(a)*(b-a))/(func(b)-func(a))
    print(iterations,'-ая итерация. Значение х=',c,'   f(x)=',func(c))
    if func(a)-func(c)<0:
        b=c
    else:
        a=c
res2=c
print('\n','Корень данного уравнения X=',c,'\n')
print(' Колличество итераций методом Хорд : ',iterations,'\n')
print(' ВЫВОД: Разность результатов методов Дихотомии и Хорд соответсвенно = ', res1-res2)

border(2)
#####*************************************************************************************************************#####
# 2)
#Выполните отделение корней графически и найдите один из корней методами простой итерации и Ньютона
#с абсолютной погрешностью 0,0001.
def func1(x):
     return 2*np.arctan(x)-x + 3

x=np.linspace(0,8,500)
plt.grid()
plt.title("График ф-ии y(x)=  2*arctan(x)-x + 3  ")
plt.plot(x,func1(x))
plt.vlines(0,-2, 4,'black')
plt.hlines(0, 0, 8,'black')
plt.show()
# Найдём произвольную точку графически . Пусть это будет х0= 6.5
# В разложение в ряд Тейлора учтём первые два слогаемые и при нуле функции , т.е. f(x0+deltx)=0
# Тогда получим ,что deltx=-f(x0)/f`(x0), где корень уравнения x1=x0+deltx
print('\t','МЕТОД НЬЮТОНА\n','\n')
x0=float(input('Введите начальное значение,опираясь на график ф-ии (x0=5) x0='))
x1=x0
# ф-ия производной
def func2(x):
    return 2/(1+x**2)-1
iterations=0
print('\n')
while abs(func1(x0))>e:
    iterations+=1
    deltx= -func1(x0)/func2(x0)
    x0+=deltx
    print(iterations,'-ая итерация.Значения Δx=',deltx,'     x=',x0,'     f(x)=',func1(x0))
res1=x0
print('\n','Корень данного уравнения X=',x0,'\n')
print(print(' Колличество итераций методом Ньютона : ',iterations,'\n'))

#МЕТОД ПРОСТОЙ ИТЕРАЦИИ
x0=x1
print('\t','МЕТОД ПРОСТОЙ ИТЕРАЦИИ \n','\n')
print('Начальное значение x0=',x0)
fx0=func2(x0)
x=np.linspace(0,8,500)
plt.grid()
plt.title("График ф-ии y(x)=  2*arctan(x)-x + 3  ")
plt.plot(x,func1(x))
plt.vlines(0, -2, 4,'black')
plt.hlines(0, 0, 8,'black')
plt.show()
# В данном методе- производная первого порядка с фиксированным значением x0 
iterations=0
while abs(func1(x0))>e:
    iterations+=1
    deltx= -func1(x0)/fx0
    x0+=deltx
    print(iterations,'-ая итерация.Значения Δx=',deltx,'     x=0',x0,'     f(x)=',func1(x0))
res2=x0
print('\n','Корень данного уравнения X=',x0,'\n')
print(' Колличество итераций методом Простой Итерации : ',iterations,'\n')
print('ВЫВОД: Разность результатов методов Ньютона и Простой Итерации соответсвенно = ', res1-res2)

border(3)
#####*************************************************************************************************************#####
#3)
# Найдите одно из решений системы уравнений методом Ньютона с абсолютной погрешностью 0,0001.
print('\t','РЕШЕНЕИЕ СИСТЕМЫ МЕТОДОМ НЬЮТОНА \n')
print('Cистема:\n','sin(x) + 2y = 2\n','cos(y − 1) + x = 0.7\n')
print('Найдём начальное приближение x0,y0 по графику ф-ий')
#ПОСТРОЕНИЕ ГРАФИКА
def Func1(x):
    return 1-0.5*np.sin(x) 
def Func2(y):
    return 0.7-np.cos(y-1)
e=0.0001
x=np.linspace(-3,4,100)
plt.grid()
plt.title("График ф-ии y(x)= sin(x) + 2y = 2 ")
plt.plot(x,Func1(x))
plt.plot(Func2(x),x)
plt.vlines(0, -3, 4,'black')
plt.hlines(0, -3, 4,'black')
plt.show()

x0=float(input('Введите начальное значение(x0=-1) x0='))
y0=float(input('Введите начальное значение(y0=2) y0='))
print('\n')
#функции для системы
def Func1_all(x,y):
    return np.sin(x) + 2*y -2
def Func2_all(x,y):
    return np.cos(y-1) + x -0.7

# Воспользуемся numpy массивами для записи матрицы
#Δx = -U(x0)*I(x0)^(-1)
X0=np.zeros((2,1),float)
X0[0,0]=x0
X0[1,0]=y0
#X0=np.array([x0,y0])
#Матрица ф-ий
#U=np.array([Func1_all(x0[0],x0[1]),Func2_all(x0[0],x0[1])])#вектор столбец
#U=np.array([Func1_all(x0[0],x0[1]),Func2_all(x0[0],x0[1])])#вектор строка
U=np.zeros((2,1),float)
U[0,0]=Func1_all(X0[0,0],X0[1,0])
U[1,0]=Func2_all(X0[0,0],X0[1,0])
#Матрица Якоби
I=np.zeros((2,2),float)
# Функции частных производных для матрицы Якоби
def Func1_all_x(x):
    return np.cos(x) 
#def func1_all_y(x,y):
#    return 2
#def func2_all_x(x,y):
#    return 1
def Func2_all_y(y):
    return -np.sin(y-1)

iterations=0

while abs(U[0,0])>e and abs(U[1,0])>e:
    iterations+=1
    I[0,0]=Func1_all_x(X0[0,0])
    I[0,1]=2
    I[1,0]=1
    I[1,1]=Func2_all_y(X0[1,0])
    
    #Обратная матрица
    I_inv=np.linalg.inv(I) 
    
    #Вектор прирощения
    #deltx=np.zeros((2,1),float)
    #deltx[0,0]=U[0,0]*I_inv[0,0]+U[1,0]*I_inv[1,0]
    #deltx[1,0]=U[0,0]*I_inv[0,1]+U[1,0]*I_inv[1,1]
    #deltx=I_inv.dot(U)
    #deltx=np.matmul(I_inv,U)
    deltx=np.dot(-I_inv,U)
    
    #ВЕКТОР-СТОЛБЕЦ ПЕРЕМЕННЫХ ПЕРЕЗАПИШЕТСЯ (+deltx)
    X0=X0+deltx
    #Матрица ф-ий перезапишется
    U[0,0]=Func1_all(X0[0,0],X0[1,0])
    U[1,0]=Func2_all(X0[0,0],X0[1,0])
    #U=np.array([Func1_all(x0[0],x0[1]),Func2_all(x0[0],x0[1])])
    print(iterations,'-ая ИТЕРАЦИЯ\n')
    print('Значение Δx=',deltx)
    print('Значение ф-ии U(X0)=',U,'\n')
    
print('РЕШЕНИЕ СИСТЕМЫ:\n','x=',X0[0,0],'\n','y=',X0[1,0],'\n')
print('Колличество итераций методом Ньютона: ',iterations)
```
