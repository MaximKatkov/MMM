```python

## Вариант No 9
### Найти точку минимума функции методом наискорейшего спуска.
### $$f(x1, x2, x3) = x_1^2 + 8x_1 + 5x_2^2 + 7x_3^2 + 119x_3 + 531.75 $$
### Для одномерной минимизации использовать метод квадратичной интерполяции.
### Для поиска интервала унимодальности использовать алгоритм скользящего окна.
##  Метод наискорейшего спуска.
### Простейшими методами первого порядка являются одношаговые методы спуска, в которых на каждом шаге вектор $p_k$ совпадает с
### направлением антиградиента функции $f(x)$:$p_k = −∇f(x_k)$.
### Такие методы называют градиентными. Градиентные методы отличаются друг от друга способом выбора множителя $α_k$.
### B методе наискорейшего спуска приближенное решение можно представить в интерационном виде $x_{k+1} = x_k + α_kp_k$, где $p_k$ - вектор, определяющий направление движения от точки $x_k$ к точке $x_{k+1}$, $α_k$ - числовой множитель, величина 
### В методе наискорейшего спуска на каждом шаге необходимо решать задачу одномерной минимизации функции переменной $α_k$ .
### Т.е. $α_k$ выбирается из условия минимума функции $f(x)$ вдоль направления $p_k$, при этом вектора $x_k$ и $p_k$ являются фиксированными . $f(x_k + α_kp_k) = min_(α<0)f(x_k + αp_k)$
### Число $α_k$ таково, что $f(x_{k+1}) = f(x_k + α_kp_k) < f(x_k)$.
### Но Точная нижняя грань функции $f_k(α) = f(x_k +αp_k)$ на полуоси $α > 0$ может и не достигаться. Но даже, если она достигается - точное определение величины $α_k$ как точки глобального на полуоси минимума функции $f_k(α)$ не всегда возможно.
### Поэтому на практике имеет смысл проводить одномерную минимизауию $f_k(α)$ на том или ином (достаточнобольшом) отрезке [0, a]. 
### В случае, когда $f_k(α)$ унимодальна на [0,а],для приближенного решения последней задачи применяют методы одномерного поиска, в данной задаче алгоритм скользящего окна
### Для выбранной исходной точки $x_0$ и выбраного окна шириной 2h > 0 около точки $x_0$ проверяктся условие унимодальности$f(x_0 − h) > f(x_0) < f(x_0 + h)$.
### Если условие выполнено то считается, что интервал унимодальности найден, в противном случае проверяется условие$f(x_0 − h) > f(x_0 + h)$.
### Если последнее выполнено, тогда окно сдвигается в право от точ-ки $x$ на $\frac{h}{2}$, иными словами точка $x$ на изменяется на точку $x =x + \frac{h}{2}$.
### В противном случае окно сдвигается в лево от точки $x$ на $\frac{h}{2}$,иными словами точка x изменяется на на точку $x = x − \frac{h}{2}$.
### Выбор ширины окна определяется экспериментально и целиком зависит от интуиции исследователя.
### Найдя окно унимодальности, можем проводить в нём одномерную минимизацию $f_k(α) = f(x_k +αp_k)$  , в данной задаче методом квадратичной интерполяцией
### Здесь задаются три пробные точки $x_1 = \frac{(a + b)}{2}$ , $x_2$ и $x_3$. Где [a,b] - является окном унимодальности.
### Для нахождения точки $x_2$ задается шаг $h > 0$ в положительном направлении от точки $x_1$, т.е. $x_2 = x_1 + h$ и если
### $f(x_1) > f(x_2)$, то $x_3 = x_1 + 2h$, иначе $x_3 = x_1 − h$.
### Далее, вычисляются значения функции в этих точках $f(x_1), f(x_2), f(x_3)$, строится квадратичный интерполяционый многочлен по трем точкам и находится его точка минимума по формуле:
### $$ x^* =\frac{1}{2} \frac{((x_2)^2 − (x_3)^2)f(x_1) + ((x_3)^2 − (x_1)^2)f(x_2) + ((x_1)^2 − (x_2)^2)f(x_3)}{(x_2 − x_3)f(x_1) + (x_3 − x_1)f(x_2) + (x_1 − x_2)f(x_3)}$$
### Находится также точка $x_{min} = arg_{min}(f(x_1), f(x_2), f(x_3))$.
### Если знаменатель в формуле для нахождения минимума квадратичного интерполяционного многочлена равен нулю, т.е. все три точки лежат на одной прямой рекомендуется выбрать за $x1 = x_{min}$ и повторить нахождение точки $x^∗$.
### Критерием окончания в этого процесса является выполнение условий для заданного e
### $|f(x_{min}) − f(x^∗)| < e,  |x_{min} − x^∗| < e$ , где е - заданная точность приближенного решения.
### Если условия окончания не выполняются и $x^∗ ∈ [x_1, x_3]$ точка $x_1$ заменяется на точку $arg_{min}(f(x_{min}), f(x^∗))$, в противном случае точка $x_1$ заменяется на $x^∗$.
### В итоге найденное значение $x^*$ и будет являться $α_{min}$ на данном шаге
### Тогда, найдя направление шага $p_k$ и его длину $α_{min}$ , с помощью интерационной формулы $x_{k+1} = x_k + α_kp_k$ 
### можем проверить приближенное значение функции на значение глобального минимума условием :
### $|f(x_{k+1}) − f(x_k| < e$ ,либо  $|x_{k+1}− x_k| < e$ 
### Если условия окончания не выполняются , то повторяем процедуру .
### Ниже представлен пример реализации метода .

import numpy as np
import matplotlib.pyplot as plt
from sympy import *


#### Вспомогательные функции для реализации методов:

def Function ( x ):
    
    return ( x[0][0]**(2) + 8.*x[0][0] + 5.*x[1][0]**(2) + 7.*x[2][0]**(2) + 119.*x[2][0] + 531.75 )


#### Функция для нахождения направления шага

def P_Direction ( X ):
    
    x1, x2, x3 = symbols('x1 x2 x3') 
    
    df_x1 = Function ( [[x1],[x2],[x3]] ).diff(x1).evalf(subs={ 'x1': X[0][0] })
    df_x2 = Function ( [[x1],[x2],[x3]] ).diff(x2).evalf(subs={ 'x2': X[1][0] })
    df_x3 = Function ( [[x1],[x2],[x3]] ).diff(x3).evalf(subs={ 'x3': X[2][0] })
    
    Norm_gradient = (df_x1**2 + df_x2**2 + df_x3**2 )**(0.5) 
     
    P = np.array( [ [ df_x1/Norm_gradient] , [df_x2/Norm_gradient] , [df_x3/Norm_gradient] ] )
    
    return P


#### Функция для нахождения градиента функции

def Gradient ( X ):
    
    x1, x2, x3 = symbols('x1 x2 x3') 
    
    df_x1 = Function ( [[x1],[x2],[x3]] ).diff(x1).evalf(subs={ 'x1': X[0][0] })
    df_x2 = Function ( [[x1],[x2],[x3]] ).diff(x2).evalf(subs={ 'x2': X[1][0] })
    df_x3 = Function ( [[x1],[x2],[x3]] ).diff(x3).evalf(subs={ 'x3': X[2][0] })
    
    gradient = np.array ([ [df_x1], [df_x2], [df_x3] ])
    
    return gradient


#### Функция для нахождения Евклидовой нормы вектора

def norm (x):
    
    return ( x[0][0]**(2)+x[1][0]**(2)+x[2][0]**(2) )**(0.5)


#### Функция графического отображения результата 

def fig_Create ( index_place , const , x0 , x , steps_x):
    
    fig = plt.subplots(1)
    
    x1 = np.arange(-3*np.pi, 3*np.pi, 0.2)
    x2 = np.arange(-3.2*np.pi, 3.2*np.pi, 0.2)
    x1grid, x2grid = np.meshgrid(x1, x2)
    
    if index_place == 0 :
        X = [[const],[x1grid],[x2grid]]
        Fgrid = Function ( X )
        plt.scatter(x[1][0],x[2][0],color='red')
        plt.scatter(x0[1][0],x0[2][0],color='blue')
        steps_x_0 = [x0[1][0]]
        steps_y_0 = [x0[2][0]]
        
        for i in steps_x:
            steps_x_0.append(i[1][0])
            steps_y_0.append(i[2][0])
        
        plt.plot(steps_x_0, steps_y_0, color = 'orange')   
        
    elif index_place == 1 :
        X = [[x1grid],[const],[x2grid]]
        Fgrid = Function ( X )
        plt.scatter(x[0][0],x[2][0],color='red')
        plt.scatter(x0[0][0],x0[2][0],color='blue')
        steps_x_0 = [x0[0][0]]
        steps_y_0 = [x0[2][0]]
        
        for i in steps_x:
            steps_x_0.append(i[0][0])
            steps_y_0.append(i[2][0])
            
        plt.plot(steps_x_0, steps_y_0, color = 'orange')  
        
    elif index_place == 2:
        X = [[x1grid],[x2grid],[const]]
        Fgrid = Function ( X )
        plt.scatter(x[0][0],x[1][0],color='red')
        plt.scatter(x0[0][0],x0[1][0],color='blue')
        steps_x_0 = [x0[0][0]]
        steps_y_0 = [x0[1][0]]
        
        for i in steps_x:
            steps_x_0.append(i[0][0])
            steps_y_0.append(i[1][0])
            
        plt.plot(steps_x_0, steps_y_0, color = 'orange')
        
    plt.contour(x1grid,x2grid, Fgrid)
    
    return plt.show()


#### Функция поска интервала унимодальности (a,b) методом скользящего окна

def Sliding_window ( alpha, x, P ) :
    
    h = 0.001
    
    while True :
    
        x1 = x - ( alpha - h/2 )* P
        x2 = x - ( alpha + h/2 )* P
        x0 = x - alpha * P
        
        if Function (x1) > Function (x0) < Function (x2):
            left = alpha - h / 2
            right = alpha + h / 2
            break
            
        elif Function (x1) > Function (x0) > Function (x2): 
            alpha += h/2
        else:
            alpha -= h/2
    
    return left , right

#### Функция поиска длины шага методом квадратичной интерполяции

def get_alpha ( x, P, left, right ):
    
    x1 = left
    x3 = right
    x2 = ((left + right)/2)
    
    arg_1 = x + left * P
    arg_2 = x + right * P
    arg_3 = x + ((left + right)/2) * P
    
    f_1 = Function ( arg_1 )
    f_2 = Function ( arg_2 )
    f_3 = Function ( arg_3 )
    
    F = { f_1 : x1 , f_2 : x2 , f_3 : x3 }
    
    key_alpha_min = min(F.keys())
    
    alpha_min = F.get( key_alpha_min )
    
    alpha_step = 0.5 * ( ( ( x2**2 - x3**2 )*f_1  +  ( x3**2 - x1**2 )*f_2  +  ( x1**2 - x2**2 )*f_3 ) )
    
    denominator = ( x2 - x3 )*f_1 + ( x3 - x1 )*f_2  + ( x1 - x2 )*f_3
    
    if denominator == 0 :
        
        left = alpha_min
        alpha_step = get_alpha( x, P, left, right )
   
    else:
        alpha_step /= denominator
    
    return alpha_step

#### Функция для нахождения приближенного значения минимума функции методом наискорейшего спуска

def gradient_Descent( x, e):
    
    counter = 0
    x_list=[]
    alpha = 0
    while True :
        
        x0 = x
        P = P_Direction ( x )
        
        Data = Sliding_window ( alpha, x, P )
        left = Data[0]
        right = Data[1]
        
        alpha = get_alpha( x, P, left, right )
        
        x = x - alpha * P 
        
        counter += 1
        x_list.append(x)
        
        if norm(x - x0) < e :
            break
        
    return x, counter, x_list



print('\n\tМетод наискорейшего спуска\n')

print(' Для одномерной минимизации - метод квадратичной интерполяции\n Для поиска интервала унимодальности - метод скользящего окна')

x0 = np.array( [ [8.] , [-2.] , [6.] ] )

result = gradient_Descent( x0, 0.0001)    

print(' Начальные координаты :\n','\nx0_1 =',x0[0][0],'\nx0_2 =',x0[1][0],'\nx0_3 =',x0[2][0],'\n')

print('Координаты минимума функции :\n','\nx1 =',result[0][0][0],'\nx2 =',result[0][1][0],'\nx3 =',result[0][2][0],'\n')

print('Колличесвто итераций :\n \nN =',result[1])

 #fig_Create ( index_place , const , x0 , x , steps_x)

fig_Create(0, result[0][0][0], x0, result[0], result[2])

fig_Create(1, result[0][1][0], x0, result[0], result[2])

fig_Create(2, result[0][2][0], x0, result[0], result[2])

### Так же, в методе наискорейшного спуска - одномерную минимизацию можно реализовать с помощью аналитического способа, приравняв первую производную функции к нулю и выразив переменную минимизации
### Это можно реализовать следующими функциями

#### Функция для нахожждения шага аналитическим методом
##### В данной функции операции производятся посимвольно и выражения записываются символами в список, после чего форматируются обратно в численный тип 

def get_Alpha(x,P):
    
    alpha, P1, P2, P3, x1, x2, x3 = symbols('alpha P1 P2 P3 x1 x2 x3')
    
    f = Function([[x1- alpha*P1] , [x2- alpha*P2] , [x3- alpha*P3]] ).diff(alpha).evalf(subs={ 'x1': x[0][0] , 'x2': x[1][0], 'x3': x[2][0], 'P1': P[0][0], 'P2': P[1][0], 'P3': P[2][0] })                               
    f = list(str(f))

    digitels_list = []
    
    for i in range(len(f)):
        ch = f.pop(0)
        if ch.isdigit():
            digitels_list.append((ch))
        elif ch == '.':
            digitels_list.append('.')
            continue
        else:
            digitels_list.append(' ')
            continue
    
    index_left = 0
    for i in range(len(digitels_list)) :
        
        if digitels_list[i] == ' ':
            index_left = i
            break
            
    index_right = 0
    for i in range(len(digitels_list)-1,0,-1) :
        
        if digitels_list[i] == ' ':
            index_right = i
            break
            
    a1 = float(''.join(digitels_list[:index_left]))
   
    b1 = float(''.join(digitels_list[index_right:]))
    
    alpha = b1 / a1
    return alpha


#### Функция для нахождения приближенного значения минимума функции методом наискорейшего спуска

def gradient_descent( x, e):
    
    counter = 0
    x_list=[]
    while True :
        x0 = x
        
        P = P_Direction ( x )
        
        alpha = get_Alpha(x,P)
        
        x = x - alpha * P 
        counter += 1
        x_list.append(x)
        
        if norm(x - x0) < e :
            break
        
    return x, counter, x_list

print('\n\tМетод наискорейшего спуска\n')

print('Для одномерной минимизации - аналитичисекий метод')

result = gradient_descent( x0, 0.0001)  

print('Начальные координаты :\n','\nx0_1 =',x0[0][0],'\nx0_2 =',x0[1][0],'\nx0_3 =',x0[2][0],'\n')

print('Координаты минимума функции :\n','\nx1 =',result[0][0][0],'\nx2 =',result[0][1][0],'\nx3 =',result[0][2][0],'\n')

print('Колличесвто итераций :\n \nN =',result[1])

 #fig_Create ( index_place , const , x0 , x , steps_x)

fig_Create(0, result[0][0][0], x0, result[0], result[2])

fig_Create(1, result[0][1][0], x0, result[0], result[2])

fig_Create(2, result[0][2][0], x0, result[0], result[2])

# Метод сопряженных градиентов 
### Метод сопряженных градиентов относится к методам первого порядка и реализуется следующим образом
### В этом методе $α_k$ выбирается, как и в методе наискорейшего спуска, из условия
### $ f(x_k + α_kp_k) = min_{α>0}f(x_k + αp_k)$
### При этом вектор $p_k$ зависит не только отградиента функции $f(x_k)$, но и от градиента в предыдущей точке$∇f(x_{k−1})$ 
### (т.е. метод является двухшаговым), и строится либо поправилу:
### $p_k = −∇f(x_k) + β_kp_{k−1}$,
### $β_0 = 0, β_k = \frac{(∇f(x_k) , ∇f(x_k)−∇f(x_{k-1})}{(∥∇f(x_{k-1}∥)^2} , k = 1, 2, . . . ; $
### переход из точки xk в точку xk+1 выполняется как в методе наискорейшего спуска
### В данном методе воспользуемся уже известными функиями нахождения интервала унимодальности и одномерной минимизации - методами скользящего окна и квадратичной интерполяции , соответсвенно

#### Функция для нахождения приближенного значения минимума функции методом сопряженного градиента

def conjugate_gradient_Descent( x, e):
    
    counter = 0
    x_list=[]
    P = - Gradient(x)
    alpha = 0
    while True :
        
        x0 = x
        P0 = P
        if counter == 0 :
            b = 0
        else:
            b =(Gradient(x)*(Gradient(x)-Gradient(x0))) / (Gradient(x0)**2)
            P = -Gradient(x) + b*P0
            
        Data = Sliding_window ( alpha, x, P )
        left = Data[0]
        right = Data[1]
        
        alpha = get_alpha( x, P, left, right )
        
        x = x - alpha * P 
        
        counter += 1
        x_list.append(x)
        
        if norm(x - x0) < e :
            break
        
    return x, counter, x_list

print('Метод сопряженных градиентов.\n')

result = conjugate_gradient_Descent( x0, 0.0001) 

print('Начальные координаты :\n','\nx0_1 =',x0[0][0],'\nx0_2 =',x0[1][0],'\nx0_3 =',x0[2][0],'\n')

print('Координаты минимума функции :\n','\nx1 =',result[0][0][0],'\nx2 =',result[0][1][0],'\nx3 =',result[0][2][0],'\n')

print('Колличесвто итераций :\n \nN =',result[1])

 #fig_Create ( index_place , const , x0 , x , steps_x)

fig_Create(0, result[0][0][0], x0, result[0], result[2])

fig_Create(1, result[0][1][0], x0, result[0], result[2])

fig_Create(2, result[0][2][0], x0, result[0], result[2])

# Градиентный метод с дроблением шага
### Градиентный метод с дроблением шага относится к методам первого порядка и реализуется следующим образом
### В данном методе значения параметров выбираются наперед, они одни и те же для всех итераций.
### Параметрами метода являются величины : длина шага - $α > 0$ ,коэффициенты дробления - $λ ∈ (0, 1) и ϵ ∈ (0, 1)$.
### Для того, чтобы выбор αk гарантировал суще-ственное убывание функции $f(x)$ при переходе от точки $x_k$ к точке $x_{k+1}$ стараются удовлетворить неравенству $f(x_{k+1}) = f(x_k + α_kp_k) < f(x_k) − ϵα_k∥∇f(x_k)∥^2$.
### Выбор αk происходит так : положив сначала $α_{k+1} = α_k$, проверим неравенство $f(x_k + α_{k+1}p_k) < f(x_k) − ϵα_{k+1}∥∇f(x_k)∥^2$.
### Если оно выполнено, то берем $α_k = α_{k+1}$. В противном случае производим постепенное дробление $α_{k+1}$, домножая его на $λ$,до тех пор, пока не выполнится условие $f(x_k + α_(k+1)p_k) < f(x_k) − ϵα_(k+1)∥∇f(x_k)∥^2$    значение $α_{k+1}$, при котором это произойдет впервые, выберем в качестве $α_k$.

#### Функция для нахождения приближенного значения минимума функции градиентным методом с дроблением шага

def steps_crush_gradient_descent( x, e):
    
    alpha = 1 
    w = 0.1
    denominator = 2
    X_steps = []
    counter = 0
    
    while True:
        
        x0 = x
        p = P_Direction ( x0 )
        
        x = x0 - alpha*p
        if norm(x - x0) < e:
            break
            
        if Function ( x ) < ( Function ( x0 ) - w*alpha*(norm(Gradient ( x0 )))*2 ) :
            X_steps.append (x)
            counter +=1
        else:
            alpha /= denominator
            
    return x, counter, X_steps

print('Градиентный метод с дроблением шага.\n')

result = steps_crush_gradient_descent( x0, 0.0001) 

print('Начальные координаты :\n','\nx0_1 =',x0[0][0],'\nx0_2 =',x0[1][0],'\nx0_3 =',x0[2][0],'\n')

print('Координаты минимума функции :\n','\nx1 =',result[0][0][0],'\nx2 =',result[0][1][0],'\nx3 =',result[0][2][0],'\n')

print('Колличесвто итераций :\n \nN =',result[1])

 #fig_Create ( index_place , const , x0 , x , steps_x)

fig_Create(0, result[0][0][0], x0, result[0], result[2])

fig_Create(1, result[0][1][0], x0, result[0], result[2])

fig_Create(2, result[0][2][0], x0, result[0], result[2]) 

# Метод Ньютона 
### Метод Ньютона относится к методам второго поряка и реализуется следующим образом
### Построим для функции$f(x)$ в окрестности точки $x_k$ формулу тейлора $f(x)= Q_k(x) + o(∥x − x_k∥^2)$,при  $∥x − x_k∥ → 0$ , где $Q_k(x)$квадратичная функция $Q_k(x) = f(x_k) + (∇f(x_k), (x−x_k)) + \frac{1}{2}(∇^2f(x_k)(x−x_k), (x−x_k))$.
### В случае, когда гессиан $∇^2f(x_k)$ положительно определен, квадратичная функция $Q_k(x)$ достигает глобального минимума в точке $x_{k+1} = x_k − [∇^2f(x_k)]^{−1}∇f(x_k)$, это и есть интерационная формула Метода Ньютона
### Если точка $x_{k+1}$достаточно близка к $x_k$, то в силу формулы $Q_k(x)$  справедливо неравенство $f(x_{k+1}) < f(x_k)$,т.е. $x_{k+1}$ естественно взять следующим за $x_k$ приближением к решению задачи
### Таким образом, в методе Ньютона $α_k = 1$, а вектор направления шага определяется решением системы линейных уравнений $[∇^2f(x_k)]p_k = −∇f(x_k)$
### Важно,что для квадратичной функции $f(x)$ метод Ньютона сходится за один шаг, что продемонстрированно в следующем примере.

#### Функция для определения направления шага в методе второго порядка, с соответсвующим нахождением матрицы Гессе и градинта .

def Newtons_P_Direction (x):
    
    # частные производные первого порядка
    
    dif_x1 = 2.*x[0][0] + 8.
    dif_x2 = 10.*x[1][0]
    dif_x3 = 14.*x[2][0] + 119.
    
    # частные производные второго порядка
    
    dif_x_11 = 2.
    dif_x_22 = 10.
    dif_x_33 = 14.
    dif_x_12 = 0  
    dif_x_21 = 0    
    dif_x_13 = 0   
    dif_x_31 = 0   
    dif_x_23 = 0    
    dif_x_32 = 0
    
    # Матрица Гессе
    A = np.array([ [dif_x_11, dif_x_12, dif_x_13],[dif_x_21, dif_x_22, dif_x_23],[dif_x_13, dif_x_23, dif_x_33] ], dtype=np.int64)
    # Вектор-столбец градиента
    gradient = np.array([[dif_x1],[dif_x2],[dif_x3]])
 
    p = np.linalg.solve(A, gradient)
    
    return p

#### Функция для нахождения приближенного значения минимума функции методом Ньютона

def Newton(x,e):
    alpha = 1
    X_steps = []
    counter = 0
    
    while True:
        x0 = x
        p = Newtons_P_Direction( x0 )
        x = x0 - alpha * p
      
        if norm(x - x0 ) < e:
            break
            
        else:
            X_steps.append(x)
            counter += 1
            
    return x, counter, X_steps


print('Метод Ньютона.\n')

result = Newton( x0, 0.0001) 

print('Начальные координаты :\n','\nx0_1 =',x0[0][0],'\nx0_2 =',x0[1][0],'\nx0_3 =',x0[2][0],'\n')

print('Координаты минимума функции :\n','\nx1 =',result[0][0][0],'\nx2 =',result[0][1][0],'\nx3 =',result[0][2][0],'\n')

print('Колличесвто итераций :\n \nN =',result[1])

 #fig_Create ( index_place , const , x0 , x , steps_x)

fig_Create(0, result[0][0][0], x0, result[0], result[2])

fig_Create(1, result[0][1][0], x0, result[0], result[2])

fig_Create(2, result[0][2][0], x0, result[0], result[2])

# Метод Ньютона-Рафсона
### Метод Ньютона-Рафсона относится к методам второго порядка и является модификацией метода Ньютона-с помощью введения переменного множителя $α_k$ и реализуется следующим образом
### В этом методе направление спуска определяется формулой $p_k = −[∇^2f(x_k)]^{(−1)}∇f(x_k)$ . A множитель $α_k$, регулирующий длину шага, можно выбирать mак чтобы выполнялось неравенство:
### $f(x_k + α_kp_k) ≤ f(x_k) + ϵα_k(∇f(x_k), p_k)$,где $ϵ ∈(0,\frac{1}{2})$-наперед заданная постоянная,одна и та же для всех итераций.
### Алгоритм нахождения множителя αk здесь такой же, как и в градиентном методе с дроблением шага, где начальное значение $α = 1$.

#### Функция для нахождения приближенного значения минимума функции методом Ньютона-Рафсона

def Newton_Rafson(x,e):
    
    alpha = 1 
    w = 0.1
    denominator = 2
    X_steps = []
    counter = 0
    
    while True:
        x0 = x
        p = Newtons_P_Direction( x0 )
        
        x = x0 - alpha * p
      
        if norm(x - x0) < e:
            break
            
        if Function ( x ) < ( Function ( x0 ) - w*alpha*(norm(Gradient ( x0 )))*2 ) :
            X_steps.append (x)
            counter +=1
        else:
            alpha /= denominator
            
    return x, counter, X_steps


print('Метод Ньютона-Рафсона.\n')

result = Newton_Rafson( x0, 0.0001)  

print('Начальные координаты :\n','\nx0_1 =',x0[0][0],'\nx0_2 =',x0[1][0],'\nx0_3 =',x0[2][0],'\n')

print('Координаты минимума функции :\n','\nx1 =',result[0][0][0],'\nx2 =',result[0][1][0],'\nx3 =',result[0][2][0],'\n')

print('Колличесвто итераций :\n \nN =',result[1])

 #fig_Create ( index_place , const , x0 , x , steps_x)

fig_Create(0, result[0][0][0], x0, result[0], result[2])

fig_Create(1, result[0][1][0], x0, result[0], result[2])

fig_Create(2, result[0][2][0], x0, result[0], result[2])

# Метод покоординатного спуска
### Метод покоординатного спуска относится к методам нулевого порядка и реализуется следующим образом.
### Вектор направления спуска $p_k$ определяется по правилу циклического покоординатного спуска: $p_k = e_{(k−[k/n]n+1)}$,при $k = 0, 1, 2, . . . $,где [$k/n$ ]- деление нацело .
### Т.е. вектор  $p_k$ определяется координатными ортами $ej = {(0, . . . , 0, 1, 0, . . . 0)}$ ,(единица стоитна j-ом месте), $j = 1, . . .,n$ . Число $α_k ∈ (−∞,∞)$ можно определять,например, следующим способом: $f(x_k + α_kp_k) = min_{(∞<α<∞)}f(x_k + αp_k).$
### Проблемы могут возникнуть, когда линии уровня сильно вытянуты, т.е. для овражных функций. В подобной ситуации поискбыстро застревает на дне такого оврага, а если начальное приближение оказывается на оси "эллипсоида то процесс так и останется вэтой точке.
### Хорошие результаты получаются в тех случаях, когда целевая функция $f(x)$ представляет собой выпуклую функцию, что показанно в следующем примере.
### В данном методе воспользуемся уже известными функиями нахождения интервала унимодальности и одномерной минимизации - методами скользящего окна и квадратичной интерполяции , соответсвенно

#### Функция для нахождения приближенного значения минимума функции методом Покоординатного спуска

def coordinate_descent(x,e):
    
    n=3
    k = 0
    alpha = 0
    X_steps = []
    
    # координатные орты
    
    e0 = np.array( [[1.],[0.],[0.]] )
    e1 = np.array( [[0.],[1.],[0.]] )
    e2 = np.array( [[0.],[0.],[1.]] )
    
    while True:
        
        x0 = x
        if (k - (k//n)*n) == 0 :
            p = e0
        elif (k - (k//n)*n) == 1:
            p = e1
        else:
            p = e2
  
        Data = Sliding_window ( alpha, x0, p )
        left = Data[0]
        right = Data[1]
        
        alpha = get_alpha( x0, p, left, right )
        
        x = x0 - alpha * p 
        
        k += 1
        X_steps.append(x)
        
        if norm(x - x0) < e :
            break
            
    return x, k, X_steps

print('Метод покоординатного спуска.\n')

result = coordinate_descent( x0, 0.0001)  

print('Начальные координаты :\n','\nx0_1 =',x0[0][0],'\nx0_2 =',x0[1][0],'\nx0_3 =',x0[2][0],'\n')

print('Координаты минимума функции :\n','\nx1 =',result[0][0][0],'\nx2 =',result[0][1][0],'\nx3 =',result[0][2][0],'\n')

print('Колличесвто итераций :\n \nN =',result[1])

 #fig_Create ( index_place , const , x0 , x , steps_x)

fig_Create(0, result[0][0][0], x0, result[0], result[2])

fig_Create(1, result[0][1][0], x0, result[0], result[2])

fig_Create(2, result[0][2][0], x0, result[0], result[2]) 

# Метод Хука-Дживса.
### Метод Хука-Дживса относится к методам нулевого порядка и реализуется следующим образом.
### В первую очередь задается начальная точка поиска $x_k$ и начальное приращение (шаг) h. После этого начинается исследующий поиск:
### Замораживаем значения всех координат кроме 1-й, вычисляем значения функции в точках x0+h0 и x0-h0 (где x0 — первая координата точки, а h0 — соответственно значение шага по этой координате) и вычисляем значение функции $f(x(k0)+h, x(k1), .., x(kn))$ и $f(x(k0)-h, x(k1), .., x(kn))$  
### и переходим в точку с наименьшим значением функции. В этой точке замораживаем значения всех координат кроме 2-й, вычисляем значения функции в точках x1+h1 и x1-h1, и вычисляем значение функции $f(x(k0), x(k1)+h, .., x(kn))$ и $f(x(k0), x(k1)-h, .., x(kn))$   переходим в точку с наименьшим значением функции и т. д. для всех координат.
### В случае, если для какой-нибудь координаты значение в исходной точке меньше, чем значения для обоих направлений шага, то шаг h по этой координате уменьшается.После чего вновь переходим к исследуемому поиску.
### Если в процессе исследующего поиска сделан хотя бы один удачный пробный шаг, то переходим к поиску по образцу:
### После исследующего поиска мы получаем точку $x_k'$. Направление $p_k = x_k'−x_k$ определяет направление, в котором функция уменьшается.
### Поэтому проводим одномерную минимизацию функции в указанном направлении, решая задачу $f(x_k + α_kp_k) = min_{(α>0)}f(x_k + αp_k)$, так чтобы $f(x_k + α_kp_k) < f(x_k).$
### В данном методе воспользуемся уже известными функиями нахождения интервала унимодальности и одномерной минимизации - методами скользящего окна и квадратичной интерполяции , соответсвенно

#### Функция для нахождения приближенного значения минимума функции методом Хука-Дживса

def Hook_Jeeves(x,e):
   
    X_steps = []
    h = 2
    counter = 0
    alpha = 0
    
    while True:
        
        x_new1,x_new2,x_new3 = x[0][0],x[1][0],x[2][0]
        
        if Function(np.array([[x_new1 + h],[x_new2],[x_new3]])) < Function(np.array([[x_new1],[x_new2],[x_new3]])):
            x_new1 = x_new1 + h
            
        elif Function(np.array([[x_new1 - h],[x_new2],[x_new3]])) < Function(np.array([[x_new1],[x_new2],[x_new3]])):
            x_new1 = x_new1 - h
            
        else:
            h /= 4
        
        if Function(np.array([[x_new1],[x_new2 + h],[x_new3]])) < Function(np.array([[x_new1],[x_new2],[x_new3]])):
            x_new2 = x_new2 + h
            
        elif Function(np.array([[x_new1],[x_new2 - h],[x_new3]])) < Function(np.array([[x_new1],[x_new2],[x_new3]])):
            x_new2 = x_new2 - h
            
        else:
            h /= 4
            
        if Function(np.array([[x_new1],[x_new2],[x_new3 + h]])) < Function(np.array([[x_new1],[x_new2],[x_new3]])):
            x_new3 = x_new3 + h
            
        elif Function(np.array([[x_new1],[x_new2],[x_new3 - h]])) < Function(np.array([[x_new1],[x_new2],[x_new3]])):
            x_new3 = x_new3 - h
            
        else:
            h /= 4
        
        if h < 1e-5:
            break
            
        p = np.array ([[x_new1 - x[0][0]],[x_new2 - x[1][0]],[x_new3 - x[2][0]]])
        
        Data = Sliding_window ( alpha, x, -p )
        left = Data[0]
        right = Data[1]
        
        alpha = get_alpha( x, p, left, right )
        
        x0 = x
        x = x + alpha*p
        
        counter += 1
        X_steps.append(x)
        
        if abs(Function(x)-Function(x0)) < e :
            
            break
        
    return x, counter, X_steps

print('Метод Хука-Дживса.\n')

result = Hook_Jeeves( x0, 0.001) 

print('Начальные координаты :\n','\nx0_1 =',x0[0][0],'\nx0_2 =',x0[1][0],'\nx0_3 =',x0[2][0],'\n')

print('Координаты минимума функции :\n','\nx1 =',result[0][0][0],'\nx2 =',result[0][1][0],'\nx3 =',result[0][2][0],'\n')

print('Колличесвто итераций :\n \nN =',result[1])

 #fig_Create ( index_place , const , x0 , x , steps_x)
    
fig_Create(0, result[0][0][0], x0, result[0], result[2])

fig_Create(1, result[0][1][0], x0, result[0], result[2])

fig_Create(2, result[0][2][0], x0, result[0], result[2]) 
```
